### Project 2: K-Means Clustering for Customer Segmentation 
#### Objective: Use K-Means clustering to segment customers based on their purchasing habits.

##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# numpy → helps handle numbers and arrays.
# matplotlib.pyplot → for plotting the clusters.
# make_blobs() → creates fake data with groups (clusters).
# KMeans → the algorithm that finds those clusters automatically.
```

##### Step 2: Generate Synthetic Data

```python
X, y = make_blobs(
    n_samples=300,       # total 300 customers (data points)
    centers=3,           # assume 3 natural customer groups
    random_state=42,     # for consistent results
    cluster_std=1.2      # spread of each cluster
)
# simulating “customer data” — for example, features could be:
# annual income
# spending score
# The function creates 3 natural groups, so K-Means can later find them.
# X → the coordinates of customers (features).
# y → the true group labels (just for checking; K-Means won’t use them
```

##### Step 3: Create and Fit the K-Means Model

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# KMeans(n_clusters=3) → tells the model to find 3 clusters.
# .fit(X) → K-Means algorithm starts:
# Picks 3 random points as “centers.”
# Assigns every customer to the nearest center.
# Moves the centers to the average of their group.
# Repeats until centers stop moving much.
# This process is called clustering.
```

##### Step 4: Get the Cluster Labels

```python
labels = kmeans.labels_
print(labels)

# After training, each data point gets a label (0, 1, or 2).
# These labels tell you which cluster each customer belongs to.
```

##### Step 5: Get Cluster Centers

```python
centers = kmeans.cluster_centers_
print(centers)

# Returns the coordinates of each cluster’s center point (mean of all customers in that cluster).
# These are the “centroids” the algorithm found.
```

##### Step 6: Visualize the Clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X')
plt.title("K-Means Customer Segmentation")
plt.xlabel("Feature 1 (e.g. Annual Income)")
plt.ylabel("Feature 2 (e.g. Spending Score)")
plt.show()

# plt.scatter() → plots all the customer points.
# c=labels → colors each point based on its cluster.
# cmap='viridis' → gives a beautiful gradient color map.
# Second plt.scatter() draws the red X marks → the cluster centers.
# The plot shows 3 colorful groups with red centers — representing different customer segments

# 3 distinct color clusters
# 3 red “X” centers
# Each group represents customers with similar purchasing patterns.
```
